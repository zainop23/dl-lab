```python
import matplotlib.pyplot as plt
import numpy as np

# Load image
img = plt.imread('inc.jpeg')
print(img)
# Convert to grayscale
r, g, b = img[:, :, 0], img[:, :, 1], img[:, :, 2]
gray = (0.299*r + 0.587*g + 0.114*b).astype(np.uint8)

# Apply threshold
threshold = 127
binary = np.where(gray > threshold, 255, 0)

# Display images
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].imshow(img)
axes[0].set_title('Original Image')

axes[1].imshow(gray, cmap='gray')
axes[1].set_title('Grayscale Image')

axes[2].imshow(binary, cmap='gray')
axes[2].set_title(f'Binary Image (Threshold={threshold})')

for ax in axes:
    ax.axis('off')

plt.tight_layout()
plt.show()

print(f"Threshold: {threshold}")
print(f"White pixels: {np.sum(binary == 255)}")
print(f"Black pixels: {np.sum(binary == 0)}")
print(f"Percentage white: {np.sum(binary == 255)/binary.size*100:.1f}%")
```
