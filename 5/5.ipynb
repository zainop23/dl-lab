```python
import numpy as np
```

```python
#XOR ddataset
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([[0],[1],[1],[0]])

def sigmoid(x,derivative=False):
    if derivative:
        return x*(1-x)
    return 1/(1+np.exp(-x))

def initialise_params():
    np.random.seed(42)
    w1 = np.random.randn(2,4)*0.5
    b1 = np.zeros((1,4))
    w2 = np.random.randn(4,1)*0.5
    b2 = np.zeros((1,1))
    return w1,b1,w2,b2
    
```

```python
def get_loss_derivative(output,y,loss_type):
    e = 1e-8
    if(loss_type=='mse'):
        return (output-y)
    elif(loss_type=='bce'):
        return (output-y)/((output+e)*(1-output+e))
    
def train(X,y,lr=0.1,loss_type='mse',epochs=500):
    w1,b1,w2,b2 = initialise_params()
    m = X.shape[0]
    # forward pass
    for _ in range(epochs):
        z1 = np.dot(X,w1)+b1
        a1 = sigmoid(z1)
        z2 = np.dot(a1,w2)+b2
        a2 = sigmoid(z2)

        #backward pass
        dz2 = get_loss_derivative(a2,y,loss_type)*sigmoid(a2,True)
        dw2 = (1/m)* np.dot(a1.T,dz2)
        db2 = (1/m)* np.sum(dz2)

        da1 = np.dot(dz2,w2.T)
        dz1 = da1 * sigmoid(a1,True)
        dw1 = (1/m) * np.dot(X.T,dz1)
        db1 = (1/m) * np.sum(dz1)

        #update params
        w1 -= lr*dw1
        b1 -= lr*db1
        w2 -= lr*dw2
        b2 -= lr*db2
    return w1,b1,w2,b2

def get_accuracy(X,y,w1,b1,w2,b2):
    z1 = np.dot(X,w1)+b1
    a1 = sigmoid(z1)
    z2 = np.dot(a1,w2)+b2
    a2 = sigmoid(z2)
    return np.mean((a2 > 0.5)==y)*100
```

```python
for lr in [0.01,0.1,1.0]:
    w1,b1,w2,b2 = train(X,y,lr=lr,loss_type='mse')
    acc = get_accuracy(X,y,w1,b1,w2,b2)
    print(f"LR = {lr} Accuracy = {acc:.1f}%")

for loss in ['mse','bce']:
    w1,b1,w2,b2 = train(X,y,lr=0.1,loss_type=loss)
    acc = get_accuracy(X,y,w1,b1,w2,b2)
    print(f"Loss type : {loss} Accuracy: {acc:.1f}%")
```
