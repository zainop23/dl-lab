```python
import torch, torchvision
import torch.nn as nn
import torch.optim as optim
import torchvision.transforms as T
from torch.utils.data import DataLoader,Subset
```

```python
t = T.Compose([T.Resize((32,32)), T.ToTensor()])
trd = DataLoader(Subset(torchvision.datasets.MNIST('./data', train=True, transform=t, download=True), range(2000)),64,True)
ted = DataLoader(Subset(torchvision.datasets.MNIST('./data', False, transform=t, download=True),range(50)),64)
```

```python
class LeNet5(nn.Module):
    def __init__(self):
        super().__init__()
        self.net = nn.Sequential(
            nn.Conv2d(1,6,5), nn.ReLU(), nn.AvgPool2d(2),
            nn.Conv2d(6,16,5), nn.ReLU(), nn.AvgPool2d(2),
            nn.Flatten(), nn.Linear(400,120), nn.ReLU(), nn.Linear(120,84), nn.ReLU(), nn.Linear(84,10)
        )
    def forward(self,x):
        return self.net(x)
```

```python
model = LeNet5()
opt = optim.Adam(model.parameters(), lr=0.001)
crit = nn.CrossEntropyLoss()
for e in range(5):
    model.train()
    rl=0
    for x,y in trd:
        opt.zero_grad()
        o = model(x)
        l = crit(o,y)
        l.backward()
        opt.step()
        rl += l.item()
    print(f'Loss: {rl/len(trd):.2f}')

model.eval()
c=n=0
with torch.no_grad():
    for x,y in ted:
        o=model(x)
        c += (o.argmax(1)==y).sum().item()
        n += len(y)
print(f'Accuracy: {c/n*100:.2f}')
```
