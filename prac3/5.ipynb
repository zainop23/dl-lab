```python
import numpy as np
```

```python
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([[0],[1],[1],[0]])

def sigmoid(x,derivative=False):
    if derivative:
        return x*(1-x)
    return 1/(1+np.exp(-x))

def initialize_params():
    np.random.seed(42)
    w1 = np.random.randn(2,4)*0.5
    b1 = np.zeros((1,4))
    w2 = np.random.randn(4,1)*0.5
    b2 = np.zeros((4,1))
    return w1,b1,w2,b2
```

```python
def loss_derivative(output,y,loss_type):
    e = 1e-8
    if loss_type=='mse':
        return output-y
    elif loss_type=='bce':
        return (output-y)/((output+e)*(1-output+e))
```

```python
def train(X,y,lr=0.1,loss_type='mse',epochs=500):
    w1,b1,w2,b2 = initialize_params()
    m = X.shape[0]
    for e in range(epochs):
        z1 = np.dot(X,w1) + b1
        a1 = sigmoid(z1)
        z2 = np.dot(a1,w2)+b2
        a2 = sigmoid(z2)

        #backward
        dz2 = loss_derivative(a2,y,loss_type)*sigmoid(a2,True)
        dw2 = (1/m)*np.dot(a1.T,dz2)
        db2 = (1/m)*np.sum(dz2)

        da1 = np.dot(w2.T,dz2)
        dz1 = da1 * sigmoid(a1,True)
        dw1 = (1/m)*np.dot(X.T, dz1)
        db1 = (1/m)*np.sum(dz1)

        #update
        w1 -= dw1*lr
        b1 -= db1*lr
        w2 -= dw2*lr
        b2 -= db2*lr
    return w1,b1,w2,b2

def accuracy(X,y,w1,b1,w2,b2):
    z1 = np.dot(X,w1) + b1
    a1 = sigmoid(z1)
    z2 = np.dot(a1,w2) + b2
    a2 = sigmoid(z2)

    return np.mean((a2>0.5)==y)*100
```

```python
for lr in [0.01,0.1,1.0]:
    w1,b1,w2,b2 = train(X,y,lr,'mse')
    acc = accuracy(X,y,w1,b1,w2,b2)
    print(f'LR: {lr} Accuracy : {acc:.2f}%')

for loss in ['mse','bce']:
    w1,b1,w2,b2 = train(X,y,0.1,loss)
    acc = accuracy(X,y,w1,b1,w2,b2)
    print(f'Loss type : {loss} accuracy : {acc:.2f}%')
```
